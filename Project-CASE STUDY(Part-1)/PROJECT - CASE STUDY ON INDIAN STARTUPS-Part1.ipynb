```python
cities=data.CityLocation.unique()
cities
```

```python
#finding indian and foreign cities(which are written together) and changing them to Indian cities
def GenerateCityNames(city):#Generate Indian city names
    notIncluded=['Bangkok','SFO','Seattle','Dubai','US','UK','USA','Palo Alto','NY','Singapore','San Mateo','New York',
                 'Houston','California','Dallas','Global']
    if '/' in city:
        cityPair=city.split('/')
        for i in range(2):#removing spaces from cityPair
            temp=cityPair[i].strip()
            cityPair[i]=temp
        
        if cityPair[0] in notIncluded:
            return cityPair[1]
        else:
            return cityPair[0]
        
    return city#return original city name

data['City']=data.CityLocation.apply(GenerateCityNames)#new column city added
cities=data.City.unique()
print(cities)
```

```python
data.head()
```

```python
data.tail()
```

```python
#Data filtering phase over
```

```python
#storing count of startups in each city
d={}
for i in cities:
    d[i]=d.get(i,0)+ (data.City==i).sum()
print(d)
```

```python
startupcount_arr=np.array([d[i] for i in d])
print(startupcount_arr)
#sorting in descending order
startupcount_arr[::-1].sort()
print(startupcount_arr)
```

```python
#finding cities with keys for finding top 10 cities
topcities=[]
for k in range(10):
    val=startupcount_arr[k]
    for i,j in d.items():
        if j==val:
            topcities.append(i)
            print(i,':',j)
```

```python
print(topcities)#for pie chart
```

### Solution to Problem 4

```python
df=pd.read_csv(r'Dataset\startup_funding.csv')
```

```python
df.head()
```

```python
print(df.InvestmentType.unique())
```

```python
print(df.InvestmentType.isnull().sum())
```

```python
#removing row where investment type is Null and rows where AmountInUSD is Null
df=df[(df.InvestmentType.notnull()) & (df.AmountInUSD.notnull())]
```

```python
print(df.InvestmentType.isnull().sum())
```

```python
#Correcting errors in Investment Type Column
df.loc[df.InvestmentType=='PrivateEquity','InvestmentType']='Private Equity'
df.loc[df.InvestmentType=='Crowd funding','InvestmentType']='Crowd Funding'
df.loc[df.InvestmentType=='SeedFunding','InvestmentType']='Seed Funding'
```

```python
uniqueInvestmentTypes=df.InvestmentType.unique()
print(uniqueInvestmentTypes)
```

```python
#removing commas from values
def removecomma(s):
    return int(s.replace(',',''))
df.AmountInUSD=df.AmountInUSD.apply(removecomma)
```

```python
df.dtypes
```

```python
df.head()
```

```python
#making a dictionary for each investment type and printing the ans
d={}
for i in uniqueInvestmentTypes:
    d[i]=df[df.InvestmentType==i].AmountInUSD.sum()

amountFundedByType=np.array([d[i] for i in d])
#sorting in descending order
amountFundedByType[::-1].sort()
s=np.sum(amountFundedByType)
FundingType_arr=[]
for i in amountFundedByType:
    for j in d:
        if d[j]==i:
            FundingType_arr.append(j)
#creating a new array of labels for pie chart also
labels=[]
for i in FundingType_arr:
    perc=(d[i]*100)/s
    print(i,'%.2f'%perc,end='%')
    labels.append(i+' ('+str('%.3f'%perc)+'%)')
    print()
```

```python
#Visualizing using a piechart
c=['red','purple','green','violet']
patches,texts=plt.pie(amountFundedByType,counterclock=False,startangle=180,radius=1.2,colors=c,wedgeprops={'edgecolor':'black'},
                     shadow=True)
#shadow parameter:makes pie 3d
#using wedge props 
#edgecolor:adds a boundary of specified color after each slice
plt.title('Investment Type distribution on the basis of funded amount')
plt.axis('equal')
plt.legend(patches,labels,loc='best')
plt.tight_layout()
plt.show()
```

### Solution to Problem 8

```python
df=pd.read_csv(r'Dataset\startup_funding.csv')
```

```python
test=df.copy()
```

```python
test.head()
```

```python
investors=test.InvestorsName.unique()
```

```python
for i in investors:
    print(i)
```

```python
test.drop(test[test.InvestorsName.isnull()].index,inplace=True)
print(test.InvestorsName.isnull().sum())
```

```python
#Removing Undisclosed Investors
def correctInvestorsName(s):
    l=['other undisclosed investors','Other undisclosed investors','Undisclosed Investors','Undisclosed investors'
          ,'undisclosed investors']

    for i in l:
        if s.__contains__(i):#checks if i is present in string or not
            out=s.replace(i,' ')
            return out
    return s
test.InvestorsName=test.InvestorsName.apply(correctInvestorsName)
```

```python
investors=test.InvestorsName.unique()
for i in investors:
    print(i)
    
    
```

```python
def ResultantDataframe(df):
    investors=df.InvestorsName.unique()
    d={}
    for i in investors:
        freq=len(df[df.InvestorsName==i])
        temp=i.split(',')
        for j in temp:
            if j!=' ' and j!='':
                t=j.strip()
                d[t]=d.get(t,0)+freq
    
    res_df=pd.DataFrame(list(d.items()))
    res_df.columns=['Investor Name','Number of times Invested']
    #sorting the resultant dataframe
    res_df.sort_values(by='Number of times Invested',inplace=True,ascending=False)
    res_df.reset_index(drop=True,inplace=True)
    return res_df
#MAIN
res_df=ResultantDataframe(test)
res_df.head()
```

```python
#ANS
print(res_df.loc[0,'Investor Name'],'has invested maximum number of times with a count of',res_df.loc[0,'Number of times Invested'])
```
